```python
import pandas as pd
import stylecloud
```

```python
data = pd.read_csv("news_labeled.csv")
data["Label"].value_counts()
```

```python
def equally_sample(df, column, n):
    equal_df = pd.DataFrame()
    uniques = df[column].unique()
    for val in uniques:
        df1 = df.loc[df[column] == val]
        df1 = df1.sample(n=n)
        equal_df = equal_df.append(df1)
    return equal_df
```

```python
balanced_data = equally_sample(df=data, column="Label", n=2050)
balanced_data["Label"].value_counts()
```

```python
real = balanced_data.loc[balanced_data["Label"] == 1]
fake = balanced_data.loc[balanced_data["Label"] == 0]
text_real = " ".join(real["Headline"])
text_fake = " ".join(fake["Headline"])
stylecloud.gen_stylecloud(text = text_real, output_name = "vis_true_headlines.png")
stylecloud.gen_stylecloud(text = text_fake, output_name = "vis_false_headlines.png")
```
